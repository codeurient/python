```python
def ORC_RXE(row):
    PrescriptionNumber = row.get("PrescriptionNumber")
    OrderNumberEXT = row.get("OrderNumberEXT")
    CurrentID = row.get("CurrentID")
    CurrentIDstr = str(CurrentID)
    MRN = str(row.get("MRN"))[-5:]
    # MRN = str(MRN)
    ORC = {'ORC-3.1':''}
    RXE = {'RXE-15': ''}
    if PrescriptionNumber > 0:
        ORC['ORC-3.1'] = f"SMATW-{PrescriptionNumber}-{MRN}"
        RXE['RXE-15'] = f"SMATW-{PrescriptionNumber}-{MRN}"

    else:
        if len(OrderNumberEXT) > 0:
            ORC['ORC-3.1'] = f"SMATW-{OrderNumberEXT}-{MRN}"
            RXE['RXE-15'] = f"SMATW-{OrderNumberEXT}-{MRN}"
        else:
            if len(CurrentIDstr) > 0:
                ORC['ORC-3.1'] = f"SMATW-{CurrentID}-{MRN}"
                RXE['RXE-15'] = f"SMATW-{CurrentID}-{MRN}"
    return ORC, RXE
```

```python
#ORC 7.4 and RXE-1.4
def get_ORC_RXE_dates(row):
    TherapyStartedDTTM = row.get("TherapyStartedDTTM")
    PerformedDTTM = row.get('PerformedDTTM')
    TherapyEndDTTM = row.get('TherapyEndDTTM')
    ClinicalDTTM = row.get('ClinicalDTTM')

    date_obj_Therapy = datetime.strptime(TherapyStartedDTTM, "%Y-%m-%d %H:%M:%S.%f") if TherapyStartedDTTM else None
    date_obj_PerformedDTTM = datetime.strptime(PerformedDTTM, "%Y-%m-%d %H:%M:%S.%f") if PerformedDTTM else None
    date_obj_TherapyEndDTTM = datetime.strptime(TherapyEndDTTM, "%Y-%m-%d %H:%M:%S.%f") if TherapyEndDTTM else None
    date_obj_ClinicalDTTM = datetime.strptime(ClinicalDTTM, "%Y-%m-%d %H:%M:%S.%f") if ClinicalDTTM else None

    if not date_obj_TherapyEndDTTM:
        return ''
    if date_obj_Therapy is not None and date_obj_Therapy.strftime("%Y") != "1900" and date_obj_Therapy.strftime("%d") != "21":
        if date_obj_Therapy.strftime("%H:%M:%S") == "00:00:00":
            return date_obj_Therapy.strftime("%Y%m%d")
        else:
            return date_obj_Therapy.strftime("%Y%m%d%H%M%S")
    else:
        return ''

    if date_obj_PerformedDTTM is not None and date_obj_PerformedDTTM.strftime("%Y") != "1900" and date_obj_PerformedDTTM.strftime("%d") != "21":
        if date_obj_PerformedDTTM.strftime("%H:%M:%S") == "00:00:00":
            return date_obj_PerformedDTTM.strftime("%Y%m%d")
        else:
            return date_obj_PerformedDTTM.strftime("%Y%m%d%H%M%S")
    else:
        return ''

    if date_obj_TherapyEndDTTM is not None and date_obj_TherapyEndDTTM.strftime("%Y") != "1900" and date_obj_TherapyEndDTTM.strftime("%d") != "21":
        if date_obj_TherapyEndDTTM.strftime("%H:%M:%S") == "00:00:00":
            return date_obj_TherapyEndDTTM.strftime("%Y%m%d")
        else:
            return date_obj_TherapyEndDTTM.strftime("%Y%m%d%H%M%S")
    else:
        return ''

    if date_obj_ClinicalDTTM is not None and date_obj_ClinicalDTTM.strftime("%Y") != "1900" and date_obj_ClinicalDTTM.strftime("%d") != "21":
        return date_obj_ClinicalDTTM.strftime("%Y%m%d%H%M%S")
    else:
        return datetime.now().strftime("%Y%m%d%H%M%S")

```
